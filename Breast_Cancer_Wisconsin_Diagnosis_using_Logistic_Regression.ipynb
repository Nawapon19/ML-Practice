**Breast Cancer Wisconsin Diagnosis using Logistic Regression**

**Dataset** :
It is given by Kaggle from UCI Machine Learning Repository, in one of its challenge
https://www.kaggle.com/uciml/breast-cancer-wisconsin-data. It is a dataset of Breast Cancer patients with Malignant and Benign tumor.
Logistic Regression is used to predict whether the given patient is having Malignant or Benign tumor based on the attributes in the given dataset.

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
# read dataset to pandas dataframe
data = pd.read_csv('data.csv')
data.head()
```

```python
# inspect the data
data.info()
```

```python
# drop column 'id' and column 'Unnamed: 32' as they have no role in prediction
data = data.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
# apply one-hot encoding to target column
# M: Malignant -> 1
# B: Benigh -> 0
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
```

```python
data
```

```python
# assign input data and label to variable x and y
y = data.diagnosis.values
x_data = data.drop(['diagnosis'], axis=1)
```

```python
# apply Min-max normalization to input data 'x'
# scales the data between 0 and 1
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
x
```

```python
# split data for training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 42)

# transpose all x and y variables
x_train = x_train.T
x_test = x_test.T
y_train = y_train.T
y_test = y_test.T

print("x train: ", x_train.shape)
print("x test: ", x_test.shape)
print("y train: ", y_train.shape)
print("y test: ", y_test.shape)
```

```python
# define a function to initialize weights and bias
def initialize_weights_and_bias(dimension):
  w = np.full((dimension, 1), 0.01)
  b = 0.0
  return w, b
```

```python
# define sigmoid function for calculating z value
def sigmoid(z):
  g_z = 1/(1 + np.exp(-z))
  return g_z
```

```python
# define a function for gradient descent calculation
def forward_backward_propagation(w, b, x_train, y_train):
  z = np.dot(w.T, x_train) + b
  g_z = sigmoid(z)
  loss = - y_train * np.log(g_z) - (1 - y_train) * np.log(1 - g_z)

  # x_train.shape[1] is for scaling
  cost = (np.sum(loss)) / x_train.shape[1]

  #backward_propagation
  derivative_weight = (np.dot(x_train, ((g_z - y_train).T))) / x_train.shape[1]
  derivative_bias = np.sum(g_z - y_train) / x_train.shape[1]
  gradients = {"derivative_weight": derivative_weight,
               "derivative_bias": derivative_bias}

  return cost, gradients
```

```python
# define a function for updating parameters
def update(w, b, x_train, y_train, learning_rate, number_of_iteration):
  cost_list = []
  cost_list2 = []
  index = []

  # update(learning) parameters number_of_iteration times
  for i in range(number_of_iteration):
    # make forward and packward propagation and find cost and gradients
    cost, gradients = forward_backward_propagation(w, b, x_train, y_train)
    cost_list.append(cost)

    # update parameter w and b
    w = w - learning_rate * gradients["derivative_weight"]
    b = b - learning_rate * gradients["derivative_bias"]
    # print cost every 10 iterations
    if i % 10 == 0:
      cost_list2.append(cost)
      index.append(i)
      print("Cost after iteration % i: % f" %(i, cost))

  # plot updates of parameter weights and bias
  parameters = {"weight": w, "bias": b}
  plt.plot(index, cost_list2)
  plt.xticks(index, rotation = 'vertical')
  plt.xlabel("Number of Itertion")
  plt.ylabel("Cost")
  plt.show()

  return parameters, gradients, cost_list
```

```python
# define a function for prediction
def predict(w, b, x_test):
  # x_test is an input for forward propagation
  z = sigmoid(np.dot(w.T, x_test) + b)
  Y_prediction = np.zeros((1, x_test.shape[1]))

  # if z is bigger than 0.5, prediction is one
  # if z is smaller then 0.5, prediction is zero
  for i in range(z.shape[1]):
    if z[0, i] <= 0.5:
      Y_prediction[0, i] = 0
    else:
      Y_prediction[0, i] = 1

  return Y_prediction
```

```python
# define a function for logistic regression model
def logistic_regression(x_train, y_train, x_test, y_test, learning_rate, num_iterations):
  # initialize parameter w and b
  dimension = x_train.shape[0]
  w, b = initialize_weights_and_bias(dimension)

  #gradient descent for updating parameter w and b
  parameters, gradients, cost_list = update(w, b, x_train, y_train, learning_rate, num_iterations)

  y_prediction_train = predict(parameters["weight"], parameters["bias"], x_train)
  y_prediction_test = predict(parameters["weight"], parameters["bias"], x_test)

  #train and test errors using mean absolute error
  print("train accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
  print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
```

```python
# train the model and get the output result
logistic_regression(x_train, y_train, x_test, y_test, learning_rate = 1, num_iterations = 150)
```
